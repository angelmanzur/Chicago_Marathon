```python
# miles per hour at every 5 km, half and full marathon
kmh_data = ['kmh05','kmh10','kmh15','kmh20','kmhHalf','kmh25','kmh30','kmh35','kmh40','kmhFinal']
times = pd.DataFrame()
for elem in kmh_data:
    rdata[elem] = pd.to_numeric(rdata[elem],errors='coerce')
```

```python
# miles per minute every 5 km, half and full marathon
mperk_data = ['mk05','mk10','mk15','mk20','mkHalf','mk25','mk30','mk35','mk40','mkFinal']
# 
pace_data = ['t05k','t10k','t15k','t20k','tHalf','t25k','t30k','t35k','t40k','tFinal']
```

```python
# convert the string data into pandas datetime 
minpermile = pd.DataFrame(columns = mperk_data)
for elem in minpermile:
    print elem,
    minpermile[elem] = rdata[elem].apply(lambda x: pd.to_datetime(x, errors='coerce', format='%M:%S'))
```

```python
#convert the pace (time per km) from string to a pandas datetime
pace = pd.DataFrame(columns = pace_data)
for elem in pace:
    print elem, 
    pace[elem] = rdata[elem].apply(lambda x: pd.to_datetime(x, errors='coerce', format='%H:%M:%S'))
```

```python
# declare function ot get the total seconds from the date
# date format is hour: minute: second
def get_seconds(date):
    hour = date.hour
    minute = date.minute
    sec = date.second
    total_seconds = hour*3600 + minute*60 + sec 
    return total_seconds

#function to convert date time into minutes (float)
# when the date format is minutes:second
def get_mins(date):
    minute = date.minute
    sec = date.second
    total_mins =  minute + sec/60. 
    return total_mins
```

```python
#pace_sec is the time in seconds passed since the race started
pace_sec = pd.DataFrame(columns = pace_data)

for elem in pace:
    print elem, 
    pace_sec[elem] = pace[elem].apply(lambda x: get_seconds(x))

print pace['t05k'][0]
aaa =  get_seconds(pace['t05k'][0])
```

```python
minpmile = pd.DataFrame(columns = mperk_data)
for elem in minpmile:
    print elem,
    minpmile[elem] = minpermile[elem].apply(lambda x: get_mins(x))
```
