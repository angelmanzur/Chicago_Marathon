```python
data_path = 'Data/RunnerData/'
nrun = 2
run_file = data_path + 'RunnerBIB_%05d.csv' %data.BIB[nrun]
print 'opening file -> %s' %run_file
rdata = pd.read_csv(run_file)
#remove the unnamed element
rdata.pop('Unnamed: 0')
#rdata['BIB'] = data.BIB[nrun]
rdata.set_index([[nrun]],inplace=True)
rdata.head(1)
#for each element, add them to the data dataframe.
```

```python
data.head()
```

```python
# add each element to the original data
for elem in rdata:
    data[elem] = ''
```

```python
#loop over all runners, 
#look at the BIB number, and load the dataframe for that BIB to rdata
#pop the unnamed
#add the values from rdata to data for that entry
#after looping, save the data to csv and exit. BOOM
starttime = time.time()
localtime = starttime
nrunners = len(data.BIB)
#nrunners = 200
mod = 100
data_path = 'Data/RunnerData/'
for nf in xrange(160,nrunners):
    #read the data 
    rtemp_file = data_path + 'RunnerBIB_%05d.csv' %data.BIB[nf]
    rdata = pd.read_csv(rtemp_file)
    #remove the first element
    rdata.pop('Unnamed: 0')
    # add the info to data dataframe
    for elem in rdata:
        data[elem][nf] = rdata[elem][0]
    if nf%mod ==0 and nf>1:
        prevtime = localtime
        localtime = time.time()
        tpf = (localtime-prevtime)/(mod)
        print '%d Files  \t %s, nfiles/sec: %1.2f (%1.2f mins -> left %1.2f min)' %(nf,rtemp_file,tpf, (localtime-starttime)/60., (nrunners-nf)*tpf/60.)
        data.to_csv('Results_allinfo.csv')
print 'Done scanning through files'
localtime = time.time()
print 'Total Time %1.3f mins' %((localtime-starttime)/60.)
data.to_csv('Results_allinfo.csv')
```
